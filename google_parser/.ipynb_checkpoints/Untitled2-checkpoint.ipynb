```python
import os
import json

# Замените на свои пути
folders = [
    "/Users/samoilov2004/Desktop/google_parser/output",
    "/Users/samoilov2004/Desktop/google_parser/output_v4_parser_v2"
    # Можно добавить ещё пути, если нужно
]

all_records = []

for folder in folders:
    for filename in os.listdir(folder):
        if filename.endswith('.json'):
            filepath = os.path.join(folder, filename)
            with open(filepath, "r", encoding="utf-8") as f:
                data = json.load(f)
                if isinstance(data, list):
                    all_records.extend(data)
                else:
                    all_records.append(data)

# Сохраним всё в общий файл
with open("all_patents_big_dick.json", "w", encoding="utf-8") as f:
    json.dump(all_records, f, ensure_ascii=False, indent=2)

print(f"Итого объединено {len(all_records)} патентов.")
```

```python
import re

sub_5 = "₅"
sub_0 = "₀"
digits = r'(?:50|₅₀)'
pattern_str = (
    r'(?:'
    r'\bIC' + digits + r'\s*\(\s*nM\s*\)'
    r'|\bEC' + digits + r'\s*\(\s*nM\s*\)'
    r'|\bKi\s*\(\s*nM\s*\)'
    r'|\bKd\s*\(\s*nM\s*\)'
    r')'
)
regex = re.compile(pattern_str)

# Тестовые данные
pos_examples = [
    "IC50 (nM)",
    "EC50(nM)",
    "Ki ( nM )",
    "Kd(nM)",
    f"IC{sub_5}{sub_0}(nM)",     # IC₅₀(nM)
    f"Вот IC{sub_5}{sub_0}(nM) значение",
    "IC50    (nM)",
    "Мы измерили EC50 (\t nM ) и Ki ( nM ) значения"
]
neg_examples = [
    "nonIC50data",            # part of a word, should **not** match
    "bestKi(nM)ever",         # no word boundary before Ki
    "superEC50(nM)record",    # no word boundary
    "pIC50(nM)",              # работает только с IC50(nM) как отдельным
    "xyzIC₅₀(nM)abc",         # субскрипт запихан в слово
    "IC51 (nM)",              # не 50/₅₀
]

print("--- Positive match tests ---")
for s in pos_examples:
    assert regex.search(s), f"Should match: {s}"
    print(f"OK: {s!r}")

print("\n--- Negative (should NOT match) ---")
for s in neg_examples:
    assert not regex.search(s), f"Should NOT match: {s}"
    print(f"OK: {s!r} (not matched)")
```

```python
import re

sub_5 = "₅"
sub_0 = "₀"

# IC50 или IC₅₀ или только IC, и аналогично для остальных.
digits = r'(?:50|{}{})?'.format(sub_5, sub_0)  # Опциональные цифры или субскрипты после IC/EC

pattern_str = (
    r'(?:'
        r'\bIC' + digits + r'(?:\s*[\(\)]*\s*nM\s*[\)\(]*)?'
        r'|\bEC' + digits + r'(?:\s*[\(\)]*\s*nM\s*[\)\(]*)?'
        r'|\bKi' + r'(?:\d+)?(?:\s*[\(\)]*\s*nM\s*[\)\(]*)?'   # Ki, Ki50, Ki222, Ki (nM)
        r'|\bKd' + r'(?:\d+)?(?:\s*[\(\)]*\s*nM\s*[\)\(]*)?'   # Kd, Kd50 и т.д.
    r')\b'
)

regex = re.compile(pattern_str, flags=re.IGNORECASE)



POS = [
    "IC50", 
    "EC50", 
    "IC₅₀", 
    "EC₅₀", 
    "IC50(nM)", 
    "IC₅₀ (nM)",
    "Kd(nM)", 
    "Ki (nM)", 
    "Ki",
    "Kd", 
    "EC50 nM", 
    "Ki50", 
    "Kd123", 
    "Ki222(nM)", 
    "Карточка Ki",        # рус. слова с Ki
    "Укажите IC₅₀",
    "Для IC50 значения",
    "значения Kd", 
    "\tKi50\t", 
    "\nIC50 ( nM )\n"
]
NEG = [
    "KiBind",
    "KiMax",
    "ICindex",
    "superEC50result",
    "ICdata",
    "nonIC50text",
    "ICvals",
    "myKdMeasure"
]

print("--- Должно совпасть ---")
for s in POS:
    assert regex.search(s), f"НЕ сработало: {s!r}"
    print(f"OK: {s}")

print("\n--- Не должно совпасть ---")
for s in NEG:
    assert not regex.search(s), f"ЛИШНЕЕ совпадение: {s!r}"
    print(f"OK: {s}")
```
